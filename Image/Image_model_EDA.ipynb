# üñºÔ∏è Image Data Analysis: CIFAR-10 Dataset üé®
üìö Libraries:

```python
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import cifar10
```

# üì¶ Loading the Dataset:

```python
(train_images, train_labels), (test_images, test_labels) = cifar10.load_data()
```

# üïµÔ∏è‚Äç‚ôÇÔ∏è Exploratory Data Analysis (EDA) & üé® Visualization:
# 1Ô∏è‚É£ Dataset Overview:
Getting a basic understanding of the dataset's shape and properties.

```python
print("Training data shape:", train_images.shape)
print("Test data shape:", test_images.shape)
```

# 2Ô∏è‚É£ Class Distribution:
Understanding the distribution of classes in the training set.

```python
unique, counts = np.unique(train_labels, return_counts=True)
plt.figure(figsize=(10, 5))
plt.bar(unique, counts, tick_label=['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck'])
plt.title('Class Distribution in Training Set')
plt.ylabel('Number of Images')
plt.show()
```

# 3Ô∏è‚É£ Visualizing Random Images:
Display random images from each class in the dataset.

```python
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

plt.figure(figsize=(10, 5))
for i, class_name in enumerate(class_names):
    plt.subplot(2, 5, i+1)
    class_idx = np.where(train_labels == i)[0]
    img_idx = np.random.choice(class_idx)
    plt.imshow(train_images[img_idx])
    plt.title(class_name)
    plt.axis('off')
plt.tight_layout()
plt.show()
```

# 4Ô∏è‚É£ Average Image for Each Class:
Visualize the average image for each class to get a general "template" or "feel" for each category.

```python
plt.figure(figsize=(10, 4))
for i, class_name in enumerate(class_names):
    plt.subplot(2, 5, i+1)
    class_idx = np.where(train_labels == i)[0]
    avg_img = np.mean(train_images[class_idx], axis=0).astype(np.uint8)
    plt.imshow(avg_img)
    plt.title(class_name)
    plt.axis('off')
plt.tight_layout()
plt.show()
```

# 5Ô∏è‚É£ Pixel Value Distribution:
Display the distribution of pixel values, which can be helpful in understanding the nature of images.

```python
plt.figure(figsize=(10, 5))
plt.hist(train_images.ravel(), bins=255)
plt.title('Distribution of Pixel Values in the Training Set')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.show()
```

## With these visualizations, we've gained insights into the CIFAR-10 dataset's class distribution, the average image for each class, and the general distribution of pixel values. This information can be valuable when considering preprocessing steps and model architecture decisions.
