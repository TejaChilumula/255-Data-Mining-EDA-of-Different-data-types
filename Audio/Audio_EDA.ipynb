The Free Spoken Digit Dataset (FSDD) is a simple audio/speech dataset consisting of recordings of spoken digits in English. Different speakers have recorded the digits, which offers variability in terms of accents, tonality, and other speech aspects.

# ðŸ•µï¸â€â™‚ï¸ Exploratory Data Analysis (EDA) & ðŸŽ¨ Visualization:
# 1ï¸âƒ£ Dataset Overview:
Understanding the basic structure and properties of the dataset.

```python
data_dir = './free-spoken-digit-dataset-master/recordings/'
audio_files = os.listdir(data_dir)
print(f"Number of audio files: {len(audio_files)}")
```

# 2ï¸âƒ£ Audio Waveform:
Visualize the waveform of a random audio sample to understand its temporal structure.

```python
sample_file = data_dir + np.random.choice(audio_files)
y, sr = librosa.load(sample_file, sr=None)

plt.figure(figsize=(10, 4))
librosa.display.waveshow(y, sr=sr)
plt.title('Waveform of an Audio Sample')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```

# 3ï¸âƒ£ Spectrogram:
Displaying the spectrogram of an audio sample helps to understand the frequency components over time.

```python
D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)

plt.figure(figsize=(10, 4))
librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram')
plt.show()
```

# 4ï¸âƒ£ Mel-Frequency Cepstral Coefficients (MFCCs):
MFCCs capture the timbral texture of the audio, which makes them highly useful for speech and audio processing tasks.

```python
mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)

plt.figure(figsize=(10, 4))
librosa.display.specshow(mfccs, sr=sr, x_axis='time')
plt.colorbar()
plt.title('MFCC')
plt.tight_layout()
plt.show()
```

The Free Spoken Digit Dataset (FSDD) provides a simple yet effective platform for understanding and experimenting with audio data. The visualizations give insights into the audio's waveform, frequency components, and timbral characteristics. This understanding is foundational for tasks such as audio classification, speech recognition, and more.
