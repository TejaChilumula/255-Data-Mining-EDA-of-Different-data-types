Data Visualization 1: Sample Frames from Videos

```python
cap = cv2.VideoCapture('path_to_sample_video') # Update with the path to your sample video
count = 0
while(cap.isOpened() and count < 5):
    ret, frame = cap.read()
    if ret:
        plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
        plt.show()
        count += 1
    else:
        break
cap.release()
```

üñºÔ∏è This code snippet displays a few sample frames from the video dataset, providing a visual representation of the data points for further analysis.

Data Visualization 2: Video Duration Distribution

```python
durations = [10, 15, 20, 25, 30]  # Example list of video durations
plt.hist(durations, bins=5, color='skyblue', alpha=0.7)
plt.xlabel('Duration (seconds)')
plt.ylabel('Frequency')
plt.title('Video Duration Distribution')
plt.show()
```

üï∞Ô∏è Here, we visualize the distribution of video durations using a histogram, providing insights into the spread and frequency of video lengths in the dataset.

Data Visualization 3: Action Category Distribution

```python
categories = ['Action1', 'Action2', 'Action3', 'Action4', 'Action5']  # Example list of action categories
counts = [25, 18, 30, 22, 15]  # Example corresponding counts for each category
plt.bar(categories, counts, color='lightgreen')
plt.xlabel('Action Categories')
plt.ylabel('Counts')
plt.title('Action Category Distribution')
plt.show()
```

üìä This code snippet presents a bar chart displaying the distribution of different action categories in the dataset, allowing us to understand the frequency of each action label.
